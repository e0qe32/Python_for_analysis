##Условие 3:

###Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости на основе матрицы корреляции.
###Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
```

```python
corr_matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix);
```

###Матрица корреляций показала сильную связь стоимости дома с жилой площадью (sqft_living), оценкой (grade), площадью (sqft_above). Подтвердим эти связи с помощью сводных таблиц.

###1. Связь стоимости дома и жилой площади.

```python
df.head()
```

```python
df['sqft_living_cut'] = pd.qcut(df['sqft_living'], 5, duplicates='drop')
```

```python
df['sqft_living_cut']
```

```python
df.groupby('sqft_living_cut')['price'].agg(['count', 'mean'])
```

**ВЫВОД:** Сводная таблица показала, что чем больше жилая площадь, тем больше средняя стоимость дома. Это подтверждает показания в матрице корреляций.

###2. Связь стоимости дома и оценки.

```python
df.groupby('grade')['price'].agg(['count', 'mean']).sort_values('mean')
```

**ВЫВОД:** Сводная таблица показала, что чем выше оценка дома, тем выше средняя стоимость дома. Это подтверждает показания в матрице корреляций.

###3. Связь стоимости дома и площади sqft_above.

```python
df['sqft_above']
```

```python
df['sqft_above_cut'] = pd.qcut(df['sqft_above'], 5, duplicates='drop')
```

```python
df['sqft_above_cut']
```

```python
df.groupby('sqft_above_cut')['price'].agg(['count', 'mean'])
```

**ВЫВОД:** Сводная таблица показала, что чем больше площадь sqft_above, тем больше средняя стоимость дома. Это подтверждает показания в матрице корреляций.
