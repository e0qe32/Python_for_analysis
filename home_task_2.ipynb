###5.3
Выберите дом клиенту. Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] <= 150000)][['yr_built', 'yr_renovated']]
# df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] <= 150000)]
```

Поэтому можно рассматривать три варианта: упомянутый дом 1922 года постройки и 2003 года ремонта, а также дома постройки 1981 и 1983 годов, ремонт которых не проводился.
